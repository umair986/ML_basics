```python
df_new =df.drop(['Cabin','Name','Embarked','Ticket'], axis = 1)
df_new
```

```python
total = df.isnull().sum().sort_values(ascending=False)
total
```

```python
df_new['Age']=df_new['Age'].fillna(0)
df_new['Age']=df_new['Age'].astype(int)
df_new['Age'].replace(0,df_new['Age'].mean())
```

```python
df_new['Age'].isnull().sum()
```

```python
df_new['Pclass'].isnull().sum()
```

```python
df_new['Sex'].replace(['female','male'],[0,1],inplace = True)
df_new
```

```python
Y = df_new['Survived']
Y
```

```python
X = df_new.drop(['Survived'], axis = 'columns')
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state = True)
```

```python
model = SVC()
svc = model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
import sklearn.metrics as metrics
metrics.confusion_matrix(Y_test,Y_pred)
```

```python
metrics.classification_report(Y_test,Y_pred)
```
